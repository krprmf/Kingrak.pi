## Train and eval

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming your data is stored in a DataFrame named 'df'
X = df_new.drop('incurred_amount', axis=1)
y = df_new['incurred_amount']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Create and fit the Linear Regression model
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

# Make predictions on the test set
predictions_lr = model_lr.predict(X_test)

# Evaluate the model
mse_lr = mean_squared_error(y_test, predictions_lr)
print(f'Mean Squared Error (Linear Regression): {mse_lr}')
```

```python
from sklearn.tree import DecisionTreeRegressor

# Create and fit the Decision Tree model
model_dt = DecisionTreeRegressor()
model_dt.fit(X_train, y_train)

# Make predictions on the test set
predictions_dt = model_dt.predict(X_test)

# Evaluate the model
mse_dt = mean_squared_error(y_test, predictions_dt)
print(f'Mean Squared Error (Decision Tree): {mse_dt}')
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# Scale the features for SVM
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and fit the SVM model
model_svm = SVR()
model_svm.fit(X_train_scaled, y_train)

# Make predictions on the scaled test set
predictions_svm = model_svm.predict(X_test_scaled)

# Evaluate the model
mse_svm = mean_squared_error(y_test, predictions_svm)
print(f'Mean Squared Error (SVM): {mse_svm}')
```

```python
import numpy as np

print(f'RMSE (Linear Regression): {np.sqrt(mse_lr)}')
print(f'RMSE (Decision Tree): {np.sqrt(mse_dt)}')
print(f'RMSE (SVM): {np.sqrt(mse_svm)}')
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and fit the Linear Regression model
model_lr = LinearRegression()
model_lr.fit(X_train_scaled, y_train)

# Make predictions on the test set
predictions_lr = model_lr.predict(X_test_scaled)

# Evaluate the model
mse_lr = mean_squared_error(y_test, predictions_lr)
print(f'Mean Squared Error (Linear Regression): {mse_lr}')

# Tune with L1 (Lasso) regularization
alpha_values = [0.1, 1, 10]
for alpha in alpha_values:
    model_lasso = Lasso(alpha=alpha)
    model_lasso.fit(X_train_scaled, y_train)
    predictions_lasso = model_lasso.predict(X_test_scaled)
    mse_lasso = mean_squared_error(y_test, predictions_lasso)
    print(f'RMSE (Lasso, alpha={alpha}): {np.sqrt(mse_lasso)}')

# Tune with L2 (Ridge) regularization
alpha_values = [0.1, 1, 10]
for alpha in alpha_values:
    model_ridge = Ridge(alpha=alpha)
    model_ridge.fit(X_train_scaled, y_train)
    predictions_ridge = model_ridge.predict(X_test_scaled)
    mse_ridge = mean_squared_error(y_test, predictions_ridge)
    print(f'RMSE (Ridge, alpha={alpha}): {np.sqrt(mse_ridge)}')
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Lasso
param_grid_lasso = {'alpha': [0.1, 1, 10,15,20,25]}

# Create and fit the Lasso model using GridSearchCV
grid_search_lasso = GridSearchCV(Lasso(), param_grid_lasso, scoring='neg_mean_squared_error', cv=5)
grid_search_lasso.fit(X_train_scaled, y_train)

# Get the best Lasso model
best_lasso_model = grid_search_lasso.best_estimator_

# Make predictions on the test set
predictions_lasso = best_lasso_model.predict(X_test_scaled)

# Calculate RMSE for Lasso
rmse_lasso = np.sqrt(mean_squared_error(y_test, predictions_lasso))
print(f'Best RMSE (Lasso): {rmse_lasso}')
print(f'Best parameters for Lasso: {grid_search_lasso.best_params_}')

# Define the parameter grid for Ridge
param_grid_ridge = {'alpha': [0.001, 0.1, 1, 10, 15]}

# Create and fit the Ridge model using GridSearchCV
grid_search_ridge = GridSearchCV(Ridge(), param_grid_ridge, scoring='neg_mean_squared_error', cv=5)
grid_search_ridge.fit(X_train_scaled, y_train)

# Get the best Ridge model
best_ridge_model = grid_search_ridge.best_estimator_

# Make predictions on the test set
predictions_ridge = best_ridge_model.predict(X_test_scaled)

# Calculate RMSE for Ridge
rmse_ridge = np.sqrt(mean_squared_error(y_test, predictions_ridge))
print(f'Best RMSE (Ridge): {rmse_ridge}')
print(f'Best parameters for Ridge: {grid_search_ridge.best_params_}')
```

# Power BI part

```python
power_bi = df[['Country', 'PublicationName', "Publisher",'Subject_Area','Type',"Citation_Number","Year", "Index_authors"]]
```

```python
power_bi.info()
```

```python
power_bi.Year.value_counts()
```

```python
power_bi = power_bi[(power_bi.Year >= 1970)]
power_bi = power_bi[(power_bi.Year <= 2023)]
```

```python
power_bi['Year'] = power_bi["Year"].apply(int)
```

```python
power_bi.Year.unique()
```

```python
power_bi = power_bi.dropna(subset=['Country', 'Subject_Area'])
```

```python
power_bi
```

```python
def count_commas(text):
    return text.count(',')
```

```python
power_bi["index_count"] = (power_bi['Country'].apply(count_commas)+1)*(power_bi["Subject_Area"].apply(count_commas)+1)
```

```python
power_bi["index_count"]
```

```python
power_bi['nomal_citation'] = power_bi['Citation_Number']/power_bi["index_count"]
```

```python
power_bi.head()
```

Next we use power qurey in Power BI

```python
power_bi.to_csv("Data2powerbi.csv", index =False)
```

# Gephi Part

```python
gephi = df[['Index_authors']]
gephi
```

Dropna


```python
gephi = gephi.dropna()
```

```python
gephi["Number_Author"] = gephi['Index_authors'].apply(count_commas)+1
```

```python
gephi
```

```python
box =gephi.Number_Author.value_counts()
```

```python
gephi.Number_Author.max()
```

```python
import matplotlib.pyplot as plt

list_ = []
for i, j in box.items():
  list_.append([i,j])
values, frequencies = zip(*list_)

# Create a bar chart
plt.bar(values, frequencies, color='skyblue')
plt.ylim(0, 2000)
# Add labels and title
plt.title("Co-Author")
plt.ylabel('Frequency', fontsize=12)
plt.xlabel('NUmber of co-author ', fontsize=12)
```
